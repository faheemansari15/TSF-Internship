```python
s_data = s_data.drop(columns="Country")
```

**Postal Code indirectly tells us about State and City so, we have to drop it too**

**Analysis with Correlation**

**Analysis asociated with Sales, Discount and Profit with respect to Ship Mode**

```python
s_data_ship = s_data.groupby(['Ship Mode'])[['Sales', 'Discount', 'Profit']].mean() 
s_data_ship
```

**Visualizing using Pie Plot**

```python
s_data_ship.plot.pie(subplots=True, figsize=(20,20), autopct='%1.1f%%', startangle=90)
plt.show()
```

***In the above,<br>*** 
***Profit and Discount is most in First Class and Sales is most Same Day (Ship Mode)<br>***
***Sales and Profit is least in Standrd Class and least Discount in Second Class***

**With respect to Segments**

**Visualizing the Segment**
